```python
def get_indexes(mol, atom_type):
    c_index = []
    for i in range(len(mol)):
        if mol[i].symbol == atom_type:
            c_index.append(i)
    return c_index
```

```python
#porph=ase.io.xyz.read_xyz("porph_bad.xyz", index=":")
mol=read("porph_C_carga_0.xyz")
view(mol, viewer='x3d')

#mol=read("benzene.xyz")
#for i in range(len(mol)):
#    print(mol[i])
```

```python
C_index = get_indexes(mol,"C")
(C_index)
```

```python
# Substituicoes simples (0, 1 e 2)
# 0=4,5,9,10,14,15,19
# 1=3,6,8,11,13,16,18
# 2=7,12,17

kn=0
kb=0

for i in range(len(C_index)):
    # +1 2
    mol[C_index[i]].symbol = "N"
    write(f"xyzs/porph_N{i}_carga_1.xyz", mol, format='xyz')
    kn+=1

    # -1 2
    mol[C_index[i]].symbol = "B"
    write(f"xyzs/porph_B{i}_carga_-1.xyz", mol, format='xyz')
    kb+=1
    mol[C_index[i]].symbol = "C" #back to C
```

```python
print(kn+kb)
```

```python
#OLD TODAS AS PERMUTAÇÕES - 800 estruturas

knn=0
kbb=0 
kbn = 0
knb = 0
# double change C to N
for i in range(len(C_index)):    
    for j in range(i+1):
        if i == j:
            None
        else:
            # +2 2
            mol[C_index[i]].symbol = "N"
            mol[C_index[j]].symbol = "N"
            write(f"xyzs/porph_N{i}_N{j}_carga_2.xyz", mol, format='xyz')
            knn+=1

            # -2 2
            mol[C_index[i]].symbol = "B"
            mol[C_index[j]].symbol = "B"
            write(f"xyzs/porph_B{i}_B{j}_carga_-2.xyz", mol, format='xyz')
            kbb+=1
            
            # 0 2
            mol[C_index[i]].symbol = "N"
            mol[C_index[j]].symbol = "B"
            write(f"xyzs/porph_N{i}_B{j}_carga_0.xyz", mol, format='xyz')
            kbn+=1
            
            # 0 2
            mol[C_index[i]].symbol = "B"
            mol[C_index[j]].symbol = "N"
            write(f"xyzs/porph_N{j}_B{i}_carga_0.xyz", mol, format='xyz')
            knb+=1
            mol[C_index[i]].symbol = "C"
            mol[C_index[j]].symbol = "C" #back to C

print(knn, kbb, kbn, knb)
```
